# Complete Beginner-Friendly EDA Notebook

This notebook demonstrates a **full Exploratory Data Analysis (EDA)** workflow.

> **Hint (Common Mistake):** Beginners often jump to modeling without understanding the data. EDA is always the first step.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Hint: Import all libraries at the start to avoid confusion later.
```

## Load Dataset
We create a small dataset for demonstration.

> **Hint:** Always check if your dataset loaded correctly before proceeding.

```python
data = {
    'product': ['A','B','C','D','E','F'],
    'category': ['Food','Clothes','Food','Electronics','Clothes','Food'],
    'price': [10,25,12,200,40,8],
    'quantity': [2,1,4,1,2,5],
    'discount': [0,5,np.nan,10,5,0]
}
df = pd.DataFrame(data)
df
# Hint: Printing the dataframe helps catch data issues early.
```

## Data Understanding
We inspect structure and statistics.

> **Hint:** Skipping `.info()` is a very common beginner mistake.

```python
df.info()
```

```python
df.describe()
# Hint: Look for unusually high max/min values â€” they may indicate outliers.
```

## Data Cleaning
Handle missing values and duplicates.

> **Hint:** Never ignore missing values; always justify how you handle them.

```python
df.isna().sum()
df['discount'] = df['discount'].fillna(0)
# Hint: Filling missing values without understanding context can lead to wrong conclusions.
```

```python
# Missing values
df.isnull().sum()
```

```python
# Duplicate rows
df.duplicated().sum()
```

## Feature Engineering
Create meaningful new features.

> **Hint:** Raw columns often do not answer business questions directly.

```python
df['total_sales'] = df['price'] * df['quantity'] - df['discount']
df
# Hint: Always verify new feature calculations manually for a few rows.
```

```python
# Grouping by category
df.groupby("category")["total_sales"].sum()
```

## Univariate Analysis
Analyze single variables.

> **Hint:** Plot distributions before applying transformations.

```python
sns.histplot(df['price'], bins=5)
plt.title('Price Distribution')
plt.show()
```

## Bivariate Analysis
Understand relationships.

> **Hint:** Correlation does NOT imply causation.

```python
sns.scatterplot(x='price', y='quantity', data=df)
plt.title('Price vs Quantity')
plt.show()
```

## Multivariate Analysis
Examine multiple variables together.

> **Hint:** Too many features can make plots hard to interpret.

```python
sns.heatmap(df[['price','quantity','discount','total_sales']].corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

## Final Takeaways

- EDA is iterative
- Cleaning improves reliability
- Features improve insight
- Visualization validates assumptions

> **Final Hint:** Always document your reasoning â€” EDA is as much about explanation as analysis.
