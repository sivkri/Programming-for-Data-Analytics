# EDA with Pandas + Matplotlib + Seaborn
Includes line, bar, and scatter plots using both libraries.

```python
import pandas as pd

data = {
    "product": ["A", "B", "C", "D", "E", "F"],
    "category": ["Food", "Clothes", "Food", "Electronics", "Clothes", "Food"],
    "price": [10, 25, 12, 200, 40, 8],
    "quantity": [5, 2, 8, 1, 3, 10],
    "date": [
        "2024-01-01", "2024-01-02", "2024-01-03",
        "2024-01-04", "2024-01-05", "2024-01-06"
    ]
}

df = pd.DataFrame(data)
df
```

```python
df.to_csv("sales.csv", index=False)
"sales.csv saved!" 
```

```python
# Import the file
df = pd.read_csv("sales.csv")
df.head()
```

```python
df["total"] = df["price"] * df["quantity"]
df["date"] = pd.to_datetime(df["date"])
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  # aesthetic styling
```

## Matplotlib Line Chart

```python
plt.plot(df["date"], df["total"])
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.title("Total Sales Over Time (Matplotlib)")
plt.show()
```

## Matplotlib Bar Chart

```python
plt.bar(df["product"], df["total"])
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.title("Sales by Product (Matplotlib)")
plt.show()
```

## Matplotlib Scatter Plot

```python
plt.scatter(df["price"], df["quantity"])
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.title("Price vs Quantity (Matplotlib)")
plt.show()
```

## Seaborn Line Chart

```python
sns.lineplot(x="date", y="total", data=df)
plt.title("Total Sales Over Time (Seaborn)")
plt.show()
```

## Seaborn Bar Chart

```python
sns.barplot(x="product", y="total", data=df)
plt.title("Sales by Product (Seaborn)")
plt.show()
```

## Seaborn Scatter Plot

```python
sns.scatterplot(x="price", y="quantity", data=df)
plt.title("Price vs Quantity (Seaborn)")
plt.show()
```
