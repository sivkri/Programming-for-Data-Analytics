## Data Cleaning & Preparation

```python
import pandas as pd

# Initial DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "David"],
    "Marks": ["85", "92", "78", "85", "90"],
    "Subject": ["Math", "Science", "History", "Math", "Art"]
}
df = pd.DataFrame(data)
```

```python
print("Original DataFrame:")
print(df)
```

```python
print("\nOriginal dtypes:")
print(df.dtypes)
```

```python
# 1. Type Conversion: Convert 'Marks' to integer
df["Marks"] = df["Marks"].astype(int)
print("\nDataFrame after type conversion (Marks to int):")
print(df)
```

```python
print("\nNew dtypes after type conversion:")
print(df.dtypes)
```

```python
# 2. Remove Duplicates
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicates:")
print(df_no_duplicates)
```

```python
# 3. Rename Column: Rename 'Marks' to 'Score'
df_renamed = df.rename(columns={"Marks": "Score"})
print("\nDataFrame after renaming 'Marks' to 'Score':")
print(df_renamed)
```

```python
# 4. String Operations: Convert 'Name' to uppercase
df_upper_name = df.copy() # Create a copy to avoid modifying original df
df_upper_name["Name"] = df_upper_name["Name"].str.upper()
print("\nDataFrame after converting 'Name' to uppercase:")
print(df_upper_name)
```

## Indexing, Selection & Filtering

```python
import pandas as pd   
# Create a sample DataFrame
data = {'Name': ['Arun', 'Bala', 'Charlie', 'Daniel', 'Aga'],        
	     'Age': [25, 30, 22, 35, 29],        
	     'Marks': [85, 92, 78, 95, 88]}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
#Accessing a specific value using df.loc
# df.loc[row_label, column_label]
marks_alice_loc = df.loc[0, "Marks"]
print("\nMarks of Alice (using df.loc[0, 'Marks']):")
print(marks_alice_loc)
```

```python
#Accessing a specific value using df.iloc
# df.iloc[row_index, column_index]
# In this case, 'Marks' is the 2nd column (index 2)
marks_alice_iloc = df.iloc[0, 2]
print("\nMarks of Alice (using df.iloc[0, 2]):")
print(marks_alice_iloc)
```

```python
# Filtering rows based on a condition using boolean indexing
# Select rows where 'Marks' are greater than 80
high_achievers = df[df["Marks"] > 80]
print("\nStudents with Marks greater than 80:")
print(high_achievers)
```

## Reshaping Data & Applying Functions

```python
import pandas as pd
import numpy as np

data = {'Name': ['Arun', 'Baldwin', 'Charlie', 'Arun', 'Baldwin', 'Charlie'],
        'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science'],
        'Marks': [85, 92, 78, 88, 75, 95]}
df = pd.DataFrame(data)
print("Original DataFrame (df):")
print(df)
```

```python
# Pivot the DataFrame
pivoted_df = df.pivot(index="Name", columns="Subject", values="Marks")
print("\nPivoted DataFrame (pivoted_df):")
print(pivoted_df)
```

```python
# Reset index for the pivoted DataFrame to make 'Name' a column again
pivoted_df_reset = pivoted_df.reset_index()

# Melt the DataFrame back to a long format
# id_vars="Name" keeps 'Name' as an identifier
# The remaining columns ('Math', 'Science') are melted
melted_df = pd.melt(pivoted_df_reset, id_vars="Name", var_name="Subject", value_name="Marks")
print("\nMelted DataFrame (melted_df):")
print(melted_df)
```

```python
# Apply the lambda function to create a new 'Grade' column
df["Grade"] = df["Marks"].apply(lambda x: "A" if x > 80 else "B")
print("\nDataFrame with new 'Grade' column:")
print(df)
```

## Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# --- Your DataFrame definition ---
data = {'Name': ['Arun', 'Baldwin', 'Charlie', 'Arun', 'Baldwin', 'Charlie'],
        'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science'],
        'Marks': [85, 92, 78, 88, 75, 95]}
df = pd.DataFrame(data)
# -----------------------------------------------

# Matplotlib often integrates directly with pandas plotting methods (df.plot.*)
# or uses standard plt.* functions with prepared data.

print("Generating visualizations using Matplotlib...")

# 1. Box plot: Marks distribution across different Subjects
#    We need to extract the 'Marks' data into lists grouped by 'Subject' for plt.boxplot
math_marks = df[df['Subject'] == 'Math']['Marks']
science_marks = df[df['Subject'] == 'Science']['Marks']
plot_data_bp1 = [math_marks, science_marks]
labels_bp1 = ['Math', 'Science']

plt.figure(figsize=(8, 5))
plt.boxplot(plot_data_bp1, tick_labels=labels_bp1) #
plt.title("Box Plot of Marks by Subject (Matplotlib)")
plt.xlabel("Subject Taught")
plt.ylabel("Marks Achieved")
plt.grid(True, axis='y', linestyle='--')
plt.show()


# 2. Box plot: Comparing Marks for each Name
#    We need to extract the 'Marks' data into lists grouped by 'Name'
arun_marks = df[df['Name'] == 'Arun']['Marks']
baldwin_marks = df[df['Name'] == 'Baldwin']['Marks']
charlie_marks = df[df['Name'] == 'Charlie']['Marks']
plot_data_bp2 = [arun_marks, baldwin_marks, charlie_marks]
labels_bp2 = ['Arun', 'Baldwin', 'Charlie']

plt.figure(figsize=(8, 5))
plt.boxplot(plot_data_bp2, tick_labels=labels_bp2)
plt.title("Box Plot of Marks by Student Name (Matplotlib)")
plt.xlabel("Student Name")
plt.ylabel("Marks Achieved")
plt.grid(True, axis='y', linestyle='--')
plt.show()


# 3. Bar plot: Mean Marks for each Subject, grouped by Name
#    This requires pre-processing the data in Pandas to calculate the means.

# Use pandas groupby to calculate the mean for each Name in each Subject
# and then unstack to get a wide format suitable for a grouped bar plot.
df_grouped_means = df.groupby(['Subject', 'Name']).mean().unstack()
df_grouped_means.columns = df_grouped_means.columns.droplevel(0) # Clean up column names

print("\nGrouped Mean DataFrame for Bar Plot:")
print(df_grouped_means)

# We can use the pandas built-in plotting function which leverages matplotlib
# to easily create grouped bar charts:
ax = df_grouped_means.plot(kind='bar', figsize=(10, 6), rot=0) #
plt.title("Average Marks for Each Student in Each Subject (Matplotlib/Pandas)")
plt.ylabel("Average Marks")
plt.xlabel("Subject")
plt.legend(title="Student Name")
plt.grid(True, axis='y', linestyle='--')
plt.show()
```

# seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# --- Your DataFrame definition (as provided) ---
data = {'Name': ['Arun', 'Baldwin', 'Charlie', 'Arun', 'Baldwin', 'Charlie'],
        'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science'],
        'Marks': [85, 92, 78, 88, 75, 95]}
df = pd.DataFrame(data)
# -----------------------------------------------

print("Generating visualizations...")

# 1. Box plot: Marks distribution across different Subjects
#    The x-axis label will automatically be "Subject".
plt.figure(figsize=(8, 5))
sns.boxplot(x="Subject", y="Marks", data=df)
plt.title("Box Plot of Marks by Subject")
plt.xlabel("Subject Taught") # Manual override for clarity if needed
plt.show()


# 2. Box plot: Comparing Marks for each Name (across subjects)
#    The x-axis label will automatically be "Name".
plt.figure(figsize=(8, 5))
sns.boxplot(x="Name", y="Marks", data=df)
plt.title("Box Plot of Marks by Student Name")
plt.xlabel("Student Name")
plt.show()


# 3. Bar plot: Mean Marks for each Subject, grouped by Name
plt.figure(figsize=(10, 6))
# 'hue' splits the bars based on the 'Name' column
sns.barplot(x="Subject", y="Marks", hue="Name", data=df, errorbar=None)
plt.title("Average Marks for Each Student in Each Subject")
plt.ylabel("Average Marks")
plt.show()
```

```python

import pandas as pd
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.transform import factor_cmap
from bokeh.palettes import Category10
import numpy as np

data = {'Name': ['Arun', 'Baldwin', 'Charlie', 'Arun', 'Baldwin', 'Charlie'],
        'Subject': ['Math', 'Math', 'Math', 'Science', 'Science', 'Science'],
        'Marks': [85, 92, 78, 88, 75, 95]}
df = pd.DataFrame(data)

print("Generating interactive plots using Bokeh...")


# 1. Box plot: Marks distribution across different Subjects in Bokeh

output_file("bokeh_boxplot_subjects.html")

subjects = df['Subject'].unique()
# Bokeh requires some pre-calculation for box plots if not using pandas_bokeh extension
# A simpler approach uses the underlying data structure:

# Calculate quartiles (requires manual aggregation as shown in Matplotlib example for multiple boxes)
groups = df.groupby('Subject')['Marks']
q1 = groups.quantile(q=0.25)
q2 = groups.quantile(q=0.5)
q3 = groups.quantile(q=0.75)
iqr = q3 - q1
upper = q3 + 1.5 * iqr
lower = q1 - 1.5 * iqr

p1 = figure(x_range=subjects.tolist(), height=350, title="Box Plot of Marks by Subject (Bokeh)",
            toolbar_location="right")

# Convert series to lists aligned with subjects for Bokeh
upper_vals = [upper[s] for s in subjects]
q3_vals = [q3[s] for s in subjects]
q2_vals = [q2[s] for s in subjects]
q1_vals = [q1[s] for s in subjects]
lower_vals = [lower[s] for s in subjects]

# Stems
p1.segment(subjects.tolist(), upper_vals, subjects.tolist(), q3_vals, line_color="black")
p1.segment(subjects.tolist(), lower_vals, subjects.tolist(), q1_vals, line_color="black")

# Whiskers (caps) - using small cap width via segment to emulate caps
cap_width = 0.2
p1.segment([s for s in subjects.tolist()], lower_vals, [s for s in subjects.tolist()], lower_vals, line_color="black")
p1.segment([s for s in subjects.tolist()], upper_vals, [s for s in subjects.tolist()], upper_vals, line_color="black")

# Boxes
p1.vbar(x=subjects.tolist(), top=q3_vals, bottom=q2_vals, width=0.7, fill_color="#E0BBE4", line_color="black")
p1.vbar(x=subjects.tolist(), top=q2_vals, bottom=q1_vals, width=0.7, fill_color="#957DAD", line_color="black")

p1.xaxis.axis_label = "Subject Taught"
p1.yaxis.axis_label = "Marks Achieved"
show(p1)


# 2. Grouped Bar Plot (More complex in vanilla Bokeh, easier with pandas_bokeh extension)
# This example uses the Pandas plotting backend for simplicity if you have the 'pandas_bokeh' library installed:

# NOTE: You need to install pandas_bokeh separately (pip install pandas_bokeh)
# import pandas_bokeh
# pandas_bokeh.output_notebook() # or output_file()

# Use pandas_bokeh to create the grouped bar plot easily
# p2 = df.plot_bokeh.bar(x="Subject", y="Marks", category="Name",
#                        title="Average Marks for Each Student in Each Subject (Pandas+Bokeh)")
# show(p2)
```

```python
import folium

# Create base map centered on average coordinates
m = folium.Map(
    location=[df["lat"].mean(), df["lon"].mean()],
    zoom_start=8
)

# Add points from DataFrame
for _, row in df.iterrows():
    folium.Marker(
        location=[row["lat"], row["lon"]],
        popup=row["name"]
    ).add_to(m)

# Display map in notebook
#for _, row in df.iterrows():
#    folium.CircleMarker(
#        location=[row["lat"], row["lon"]],
#        radius=15,
#        popup=row["name"],
#        fill=True
#    ).add_to(m)

m.save("mapABC.html")
```
